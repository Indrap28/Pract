```python
X = data.drop('sepsis',axis=1)
y = data['sepsis']
```

```python

y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote=SMOTE()
```

```python
x_res2,y_res2=smote.fit_resample(X,y)
```

```python
print(y.value_counts())
print(y_res2.value_counts())
```

```python
X_train,X_test,y_train,y_test = train_test_split(x_res2,y_res2,test_size=0.2,random_state=0)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(X_train)
x_test = sc.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred)
accuracy_rf
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)
```

```python
y_predlr=lr_model.predict(X_test)
accuracy_lr = accuracy_score(y_test, y_predlr)
accuracy_lr
```

```python
pip install xgboost
```

```python
from xgboost import XGBClassifier
xg_model = XGBClassifier()
xg_model.fit(X_train,y_train)
```

```python
X_test.shape
```

```python
y_predxgb=xg_model.predict(X_test)
accuracy_xg = accuracy_score(y_test, y_predxgb)
accuracy_xg
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(random_state = 0)
classifier.fit(X_train,y_train)
```

```python
decisiontree = classifier.predict(X_test)
```

```python
accuracy_dt=accuracy_score(y_test,decisiontree)
```

```python
accuracy_dt
```

```python
from sklearn.svm import SVC
svm = SVC(kernel = "linear")
svm.fit(X_train,y_train)
```

```python
svm_pred = svm.predict(X_test)
```

```python
accuracy_svc=accuracy_score(svm_pred,y_test)
```

```python
accuracy_svc
```
